## 4. '연령'을 이용해, 고객 '연령대'를 계산하고, '총구매금액'의 합이 가장 높은 2개의 '연령대'를 확인하여, 두 연령대 간의 '총구매금액'의 평균의 차이를 계산 하시오.

```python
def func1(x):
    if pd.isnull(x):
        return '미확인'
    elif x>=100:
        return '100세 이상'
    else : 
        return str(x)[0] + '0대'

df2['연령대'] = df2['연령'].apply(func1)
df2['연령대'].value_counts()
```

```python
p1=df2.pivot_table(index='연령대',values='총구매금액',aggfunc='sum').reset_index()
p1
```

```python
#'총구매금액'의 합이 가장 높은 2개의 '연령대'를 확인하여
p2=p1.sort_values(by = '총구매금액',ascending=False ).head()
p2
```

```python
# 두 연령대 간의 '총구매금액'의 평균의 차이를 계산 하시오.

avg_purchase_40s = df2[df2['연령대'] == '40대']['총구매금액'].mean()
avg_purchase_50s = df2[df2['연령대'] == '50대']['총구매금액'].mean()

# 평균의 차이 계산
abs(avg_purchase_40s - avg_purchase_50s)
```

## 6. '총구매금액'을 '총방문횟수'로 나누어, '1회방문구매평균'을 계산하고, 생성된 파생변수를 바탕으로 아래와 같이 고객등급을 분류하시오.
- '1회방문구매평균'이 10만원 이상 고객은 A / 5만원 이상은 B / 2만원 이상 C / 나머지는 D 등급으로 분류
- 분류된 고객 등급은 '고객등급'이라는 새로운 파생변수로 생성 

```python
df2['1회방문구매평균'] = df2['총구매금액'] / df2['총방문횟수']
df2.head()
```

```python
# '고객등급' 분류 함수 정의
def classify_grade(avg_purchase):
    if avg_purchase >= 100000:
        return 'A'
    elif avg_purchase >= 50000:
        return 'B'
    elif avg_purchase >= 20000:
        return 'C'
    else:
        return 'D'

# '고객등급' 파생변수 생성
df2['고객등급'] = df2['1회방문구매평균'].apply(classify_grade)
df2
```

## 7. 새로 생성된 파생변수 '고객등급'에서 '기혼'자가 가장 많은 그룹은 어디 그룹인지 확인하시오.

```python
marriage_counts = df2[df2['결혼유무'] == '기혼']['고객등급'].value_counts()

print(marriage_counts)

most_married_group = marriage_counts.idxmax()
most_married_group
```

## 8. df2 데이터에서 '주소'데이터를 바탕으로, "시" 별 '지역(시)'이라는 이름의 새로운 파생변수를 생성 하시오.
- 서울시 / 용인시 / 성남시 /수원시 /광주시 / 나머지 지역은 기타로 분류 

```python
def extract_city(address):
    if pd.isnull(address):
        return '기타'
    else:
        cities = ['서울', '용인', '성남', '수원', '광주']
        for city in cities:
            if city in address:
                return city + '시'
        return '기타'

df2['지역(시)'] = df2['주소'].apply(extract_city)
df2
```
