## 10. 계약일자(Datetime)에서 계약연도(Year)파생변수를 생성하고, 연도 별, 각 제품군(Product_Type)의 월랜탈비용(Amount_Month) 합을 계산하시오.

```python
df['Datetime_df'] = pd.to_datetime(df['Datetime']) #pd.to_datetime 날짜를 인식할 수 있게 해준다.
df.head()
```

```python
df[['Datetime','Datetime_df']]
```

```python
df['Year'] = df['Datetime_df'].dt.year
df.head()
```

```python
# 계약일자(Datetime)에서 계약연도(Year)파생변수를 생성하고, 
#연도 별, 각 제품군(Product_Type)의 월랜탈비용(Amount_Month) 
#합을 계산하시오
df.pivot_table( index= 'Product_Type',
               values = 'Amount_Month',
               aggfunc='sum',
               columns='Year').T
```
