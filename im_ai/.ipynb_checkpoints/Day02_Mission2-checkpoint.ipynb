## 2. 두 개의 값을 입력받아 두 값 사이의 10개의 데이터를 일정 간격으로 생성해, Vector형태로 출력하는 함수를 생성 

np.arange(1,20,3)

1부터 20까지 3의 간격으로 벡터 생성

```python
import numpy as np

def between_values():
    # 두 개의 값 입력받기
    value1 = float(input("첫 번째 값을 입력하세요: "))
    value2 = float(input("두 번째 값을 입력하세요: "))
    
    # 두 값 사이의 10개의 데이터를 일정 간격으로 생성
    vector = np.linspace(value1, value2, 10)
    
    return vector

vector = between_values()
print("생성된 벡터:", vector)
```

## 3. 학생의 이름과 국어점수, 영어점수, 수학점수를 입력받아 Pandas의 DataFrame형태로 구성하는 함수를 생성 

```python
import pandas as pd

def student_dataframe():    
    num_students = int(input("학생 수를 입력하세요: "))   
    data = {
        "이름": [],
        "국어": [],
        "영어": [],
        "수학": []
    }
    

    for _ in range(num_students):
        name = input("학생의 이름을 입력하세요: ")
        korean_score = int(input("국어 점수를 입력하세요: "))
        english_score = int(input("영어 점수를 입력하세요: "))
        math_score = int(input("수학 점수를 입력하세요: "))
        

        data["이름"].append(name)
        data["국어"].append(korean_score)
        data["영어"].append(english_score)
        data["수학"].append(math_score)
    

    df = pd.DataFrame(data)
    
    return df



student_dataframe()
```

## 6. df1_young 데이터와 df1 데이터의 월랜탈비용(Amount_Month)의 평균을 계산하고, 두 값의 차이를 계산 

```python
# df1_young의 Amount_Month 평균 계산
average_amount_young = df1_young['Amount_Month'].mean()

# df1의 Amount_Month 평균 계산
average_amount_all = df1['Amount_Month'].mean()

# 두 값의 차이 계산
difference = average_amount_young - average_amount_all

# 결과 출력
print(f"df1_young의 월렌탈비용 평균: {average_amount_young}")
print(f"df1의 월렌탈비용 평균: {average_amount_all}")
print(f"두 값의 차이: {difference}")
```

## 8. 본사에서 매출관리를 위해 연체가 높은 고객들을 확인하여 관리하고자 한다. 연체건수가 높은 순서대로 정렬하고, 상위 100명의 고객의 연령(Age)의 최대값과 최소값을 확인, 신용등급(Credit_Rank)의 평균을 계산 

```python
df2=df1.sort_values(by = ['Overdue_count'],
               ascending=[False])
df2.head(100)
```

```python
# 연령의 최대값과 최소값 확인
max_age = df2['Age'].head(100).max()
min_age = df2['Age'].head(100).min()
print(max_age)
print(min_age)
```

```python
# 신용등급의 평균 계산
average_credit_rank = df2['Credit_Rank'].head(100).mean()

# 결과 출력
print(f"상위 100명의 고객 연령의 최대값: {max_age}")
print(f"상위 100명의 고객 연령의 최소값: {min_age}")
print(f"상위 100명의 고객 신용등급의 평균: {average_credit_rank:.2f}")
```

## 9. 매출관리 부서 A씨는 매주 해당 데이터를 받아 계약금액이 높은 1000개의 계약명단의 평균값을 계산하고 있다. 이 업무를 자동화 하기 위해, 월랜탈비용이 높은 1000갸의 데이터를 추출하여, 해당 데이터의 월랜탈비용의 평균을 계산하는 함수를 만드시오. 

```python
top_1000 = df1.nlargest(1000, 'Amount_Month') # sorted와 동일
top_1000
```

```python
# 월렌탈비용의 평균 계산
average_amount_month = top_1000['Amount_Month'].mean()
average_amount_month
```
