## 1. df10_clean의 결제수단에는 어떤 항목이 있는지 확인한 후, 그중 '적립금'이 포함된 값들만 추출하여 df_point로 선언하시오.

```python
df10_clean['결제수단'].unique()
```

```python
def func1(row):
    if '적립금' in row:
        return True
    else : 
        return False

cond1 = df10_clean['결제수단'].apply(func1)
cond1
```

```python
df10_clean['결제수단']
```

```python
cond1.unique()
```

```python
df_point=df10_clean.loc[cond1]
df_point.head()
```

```python
df_point['결제수단'].unique()
```

## 2. 사업자와 일반 고객의 사용한 적립금액의 합과 평균을 계산한 후 각각의 차를 계산하시오

```python
df_point.head()
```

```python
cond1 = (df_point['주문 시 회원등급'] =='사업자')
cond2 = (df_point['주문 시 회원등급'] =='일반')

df_busi = df_point.loc[cond1]
df_norm = df_point.loc[cond2]

bm=df_busi['사용한 적립금액'].mean()
nm=df_norm['사용한 적립금액'].mean()

bs=df_busi['사용한 적립금액'].sum()
ns=df_norm['사용한 적립금액'].sum()

print(bm,nm,bm-nm)
```

```python
print(bs,ns,bs-ns)
```

```python
def func3():
    print(df_point['주문 시 회원등급'].unique())
    input_rank = input('회원 등급을 입력하시오')

    cond1 = df_point['주문 시 회원등급'] == input_rank
    df_point.loc[cond1]
    p1= df_point.loc[cond1]
    return p1['사용한 적립금액'].describe()
func3()
```

```python
# 숙련자 코드

def func3():
    print(df_point['주문 시 회원등급'].unique())
    input_rank = input('회원 등급을 입력하시오')
    
    if input_rank in df_point['주문 시 회원등급'].unique():
        cond1 = df_point['주문 시 회원등급'] == input_rank
        df_point.loc[cond1]
        p1= df_point.loc[cond1]
        return p1['사용한 적립금액'].describe()
    else:
        return '다시 입력하시오'
func3()
```
