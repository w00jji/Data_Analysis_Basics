# 기술 통계량 확인 ( DDA )
- 데이터의 통계적 특성을 확인
- 연속형 :
   - 대표값(평군 mean / 중앙값 median) : 해당 데이터를 대표하는값
     - 평균의 경우 이상치(outlier)에 의해 좌우될 수 있음
   - 산포 (분산 / 표준편차)
     - 대표값에 대해 데이터가 얼마나 정확한지 나타내는 척도
     - 데이터의 신뢰도를 나타내기도 함
   - 분포의 모양
     - 분포의 모양에 따라 가설검정 기법 / 회귀분석 방법이 달라질 수 있음
     - 가설검정을 통해 정규성 확인
- 범주형 : 항목 / 빈도수

# 필터
- 조건을 걸어 추출
- 데이터프레임.loc[조건식] : 특정 조건 (True)에 해당하는 데이터를 추출

```python
# 월렌탈비용 10만원 이상인 데이터들만 추출
cond1 = (df1['Amount_Month'] >=100000) 
cond1
```

```python
df1.loc[cond1]

# True에 해당하는 row만 가져온다.
```

```python
# 월렌탈비용이 10만원 이상이면서
# 연체건수 2건 이하인 데이터만 추출
cond1 = (df1['Amount_Month'] >=100000) 
cond2 = (df1['Overdue_count'] <=2) 

df1.loc[cond1 & cond2]
```

```python
df1.loc[(df1['Amount_Month'] >=100000) & (df1['Overdue_count'] <=2)] #가독성이 안좋다
```

```python
# or
df1.loc[cond1 | cond2]
```
